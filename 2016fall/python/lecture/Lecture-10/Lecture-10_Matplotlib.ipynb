## Adjusting the spacing around subplots

```python

fig, axes = plt.subplots(2, 2, sharex=True, sharey=True) 
col = ['r', 'b', 'k', 'g']
bins = [50, 40, 30, 20]
nums = randn(500)
for i in range(2):
    for j in range(2):
            axes[i, j].hist(nums, bins=bins[i +2 * j] , color = col[i + 2 * j], alpha=0.5)
plt.subplots_adjust(wspace=0, hspace=0)
```

## Ticks, Labels, and Legends

```python
fig = plt.figure() 
ax = fig.add_subplot(1, 1, 1)
ax.plot(randn(1000).cumsum())
```

```python
ticks = ax.set_xticks([0, 250, 500, 750, 1000])
labels = ax.set_xticklabels(['one', 'two', 'three', 'four', 'five'], rotation=30, fontsize='small')
ax.set_title('My first matplotlib plot')
ax.set_xlabel('Stages')
fig
```

```python
fig = plt.figure(); ax = fig.add_subplot(1, 1, 1)
ax.plot(randn(1000).cumsum(), 'b-', label='first') 
ax.plot(randn(1000).cumsum(), 'r--', label='second')
ax.plot(randn(1000).cumsum(), 'k-.', label='third')
ax.legend(loc='lower left')
```

## Annotations and Drawing on a Subplot

```python
import pandas as pd
from datetime import datetime
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
data = pd.read_csv('/Users/wuyan/Documents/courses/2016 fall/python/DataAnalysisPython/data/stock_data/data/stocks/spx.csv', index_col=0, parse_dates=True) 
spx = data['SPX']
spx.plot(ax=ax, style='k-')
crisis_data = [
  (datetime(2007, 10, 11), 'Peak of bull market'), 
    (datetime(2008, 3, 12), 'Bear Stearns Fails'), 
    (datetime(2008, 9, 15), 'Lehman Bankruptcy')
]
for date, label in crisis_data:
    ax.annotate(label, xy=(date, spx.asof(date) + 80),
    xytext=(date, spx.asof(date) + 230), 
                arrowprops=dict(facecolor='black'), 
                horizontalalignment='left', verticalalignment='top')
    # Zoom in on 2007-2010 
ax.set_xlim(['1/1/2007', '1/1/2011']) 
ax.set_ylim([600, 1800])
ax.set_title('Important dates in 2008-2009 financial crisis')
```
